```python
import matplotlib.pyplot as plt
```

```python
def plot_loss(train_loss, val_loss, epochs):
    plt.plot(range(1, epochs + 1), train_loss, 'b-', label='Training loss')
    plt.plot(range(1, epochs + 1), val_loss, 'g-', label='Validation loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
def plot_accuracy(train_acc, val_acc, epochs):
    plt.plot(range(1, epochs + 1), train_acc, 'b-', label='Training Accuracy')
    plt.plot(range(1, epochs + 1), val_acc, 'g-', label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```
